```python
 def convert3(obj):
  L = []
  counter =0
  for i in ast.literal_eval(obj):
    if counter !=3:
      L.append(i['name'])
      counter+=1
    else:
      break
  return L

```

```python
movies['cast'] = movies['cast'].apply(convert3)
```

```python
def convert3(cast):
    if isinstance(cast, list):
        # Convert list to a string, joining by a comma
        return ', '.join(cast)
    elif isinstance(cast, str):
        # If it's already a string, return as is
        return cast
    else:
        # Handle unexpected types, like None or other data types
        return ''

# Apply the function to the 'cast' column
movies['cast'] = movies['cast'].apply(convert3)
```

```python
 def fetch_director(obj):
  L = []
  for i in ast.literal_eval(obj):
    if i['job'] == 'Director':
      L.append(i['name'])
      break
  return L
```

```python
movies['crew'] = movies['crew'].apply(fetch_director)
```

```python
movies['overview'][0]
```

```python
movies['overview'] = movies['overview'].apply(lambda x:x.split())
```

```python
movies.head()
```

```python
### remove space from single words so that you get recommend from the other same name
```

```python
movies['genres'] = movies['genres'].apply(lambda x: [i.replace(" ","") for i in x] if x is not None else [])
```

```python
movies['keywords'] = movies['keywords'].apply(lambda x: [i.replace(" ","") for i in x] if x is not None else [])
movies['cast'] = movies['cast'].apply(lambda x: [i.replace(" ","") for i in x] if x is not None else [])
movies['crew'] = movies['crew'].apply(lambda x: [i.replace(" ","") for i in x] if x is not None else [])
```

```python
movies.head();
```

```python
movies['tags'] = movies['overview'] +  movies['keywords'] +  movies['crew']+  movies['cast'] +  movies['genres']
```

```python

from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
```

```python

def stem(text):
  y = []
  for i in text.split():
    y.append(ps.stem(i))
    return " ".join(y)
```

```python
new_df['tags'] = new_df['tags'].apply(stem)
```

```python

new_df['tags'][0]
```

```python
new_df['tags'] = new_df['tags'].apply(lambda x:x.lower())
```

```python
## we combine the tags (concatinate) ->large text
## we want most used words
## top words with most frequescy
##
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=5000,stop_words='english')
```

```python
vectors = cv.fit_transform(new_df['tags']).toarray()
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity = cosine_similarity(vectors)
```

```python
sorted(list(enumerate(similarity[0])),reverse=True,key=lambda x:x[1])[1:6]
```

```python
def recommend(movie):
  movie_index = new_df[new_df['title'] == movie].index[0]
  distances = similarity[movie_index]
  movies_list = sorted(list(enumerate(distances)),reverse=True,key=lambda x:x[1])[1:6]

  for i in movies_list:
    print(new_df.iloc[i[0]].title)
```

```python
recommend('The Dark Knight Rises')
```
